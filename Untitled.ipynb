```python
delivery_df.rename(columns={'total_runs_x':'total_runs','total_runs_y':'ball_scored'},inplace=True)
```

```python
delivery_df
```

```python
delivery_df['score']=delivery_df[['match_id','ball_scored']].groupby('match_id').cumsum()['ball_scored']
```

```python
delivery_df
```

```python
#new column which has target left
delivery_df['target_left']=(delivery_df['total_runs']+1)-delivery_df['score']
```

```python
delivery_df
```

```python
#trying to find out number of balls left
#1 over=6 balls
#in an ipl, 20 overs=20*6=120 balls
delivery_df['remaining_balls']=120-((delivery_df['over']-1)*6+delivery_df['ball'])
```

```python
delivery_df
```

```python
#how many wickets are left out
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else '1').astype('int64')
```

```python
delivery_df
```

```python
delivery_df['wickets']=delivery_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df
```

```python
delivery_df['wickets']=10-delivery_df['wickets']
```

```python
delivery_df['wickets']
```

current run rate (crr) = runs scored/number of overs completed

```python
delivery_df['crr']=(delivery_df['score']*6)/(120-delivery_df['remaining_balls'])
```

```python
delivery_df
```

required run rate = target/number of overs left

```python
delivery_df['rrr']=(delivery_df['target_left']*6)/(delivery_df['remaining_balls'])
```

```python
delivery_df
```

```python
#creating a column result where 1 means win and 0 means lost
def result(row):
    if row['winner']==row['batting_team']:
        return 1
    else:
        return 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
#splitting the data in features and target
x=model_df.iloc[:,:-1]
y=model_df.iloc[:,-1]
```

```python
#splitting the data in train and test parts
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
Y_train
```

```python
Y_test
```

```python
#encoding - one hot encoding or label encoding
#transform non-numerical data to numerical data -> one hot encoding
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder 
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])], remainder = 'passthrough')
```

```python
trf
```

```python
#building the model
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_prediction=pipe.predict(X_test)
```

```python
Y_prediction
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_prediction)
```
